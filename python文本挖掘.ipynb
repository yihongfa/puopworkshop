实操部分

```python
titles = df['title']#提取需分词的内容
```

```python
#按词性分词
import jieba.posseg as pseg
titles1=[]
for title in titles:
    title1=pseg.cut(title)
    title2=[[word.word,word.flag]for word in title1]
    titles1.append(title2)
```

```python
print(titles1[1])
```

```python
#按词性抽取
titles2=[[word[0] for word in text if word[1]=='n' 
         or word[1]=='v' or word[1]=='a' or word[1]=='d' or word[1]=='nr' 
         or word[1]=='nrfg' or word[1]=='ns' or word[1]=='nt' or word[1]=='nz' 
         or word[1]=='vn' or word[1]=='vd' or word[1]=='an' or word[1]=='ad' 
         or word[1]=='l']for text in titles1]
```

```python
print(titles2[1])
```

```python
#删除一个字的词
titles3 =[[word for word in text if len(word)>1]for text in titles2]
```

```python
print(titles3[1])
```

```python
len(titles3)
```

## 计算词频

```python
#合并成一个文档
text = []
for word in titles3:
    text.extend(word)
```

```python
#统计频数
from collections import defaultdict
word_dict = defaultdict(int) 
for word in text:
    word_dict[word] +=1
```

```python
word_dict
```

```python
#从字典中提取词条和词频
terms = []
freq = []
for item in word_dict.items():
    terms.append(item[0])
    freq.append(item[1])
```

```python
wordcount = pd.DataFrame({'terms':terms,'count':freq})#写入数据框
```

```python
wordcount1 = wordcount.sort_values(by='count',ascending=False)#按词频降序排列
```

```python
wordcount1
```

```python
wordcount1.to_csv('allWordsCount.csv')
```

## TF-IDF

```python
#定义用于tfidf的分词器
def tokench(text):
    tokens = [word for word in text.split(' ')]#数据以空格隔开
    after_tokens = []
    for token in tokens:
        if len(token)>1:
            after_tokens.append(token)
    return after_tokens
```

```python
#构建tfidf矩阵
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                 tokenizer=tokench,use_idf=True)

tfidf_matrix = tfidf_vectorizer.fit_transform(df['titlecut'])
```

```python
terms = tfidf_vectorizer.get_feature_names()#获取词（特征）名
```

```python
terms[:10]
```

```python
weight = tfidf_matrix.todense().tolist()#tfidf权重
```

## 文本相似度

```python
#计算距离
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(tfidf_matrix)
dist = 1.0 - cosine_similarity(tfidf_matrix)
```

## K-means 聚类

```python
#利用scikit-learn中的Kmeans模型进行聚类
from sklearn.cluster import KMeans

num_clusters = 9

km = KMeans(n_clusters=num_clusters)

km.fit(tfidf_matrix)

clusters = km.labels_.tolist()
```

```python
clusters[:10]#前10个文档的类
```

```python
df['cluster'] = clusters
```

## 主题建模

```python
#文本准备
from gensim import corpora, models, similarities
texts = titles3
```

```python
dictionary = corpora.Dictionary(texts) #构造词典
dictionary.filter_extremes(no_below=1, no_above=0.8)#去高频词

corpus = [dictionary.doc2bow(text) for text in texts] #构造语料库
```

```python
[s for s in corpus]
```

```python
#训练一个LDA模型
%time lda = models.LdaModel(corpus, num_topics=9, id2word=dictionary, update_every=5, alpha=0.1, eta=0.1)
```

```python
print(lda[corpus[0]])#打印第一个文档的主题建模结果
```

**预测文档对应的主题**

```python
doctop = [ s for s in lda[corpus]]
```

```python
doctop = pd.DataFrame(doctop)
```

```python
doctop.to_csv('DocumentTopic.csv')
```
